```python
import google.generativeai as genai

# Inicializar modelo
model = genai.GenerativeModel("gemini-3-flash-preview")

BASE_PROMPT = """
Actúa como asistente experto en organización personal adaptativa.

Organiza una semana equilibrada considerando:

HORARIOS FIJOS:
{horarios_fijos}

HORARIOS FLEXIBLES:
{horarios_flexibles}

HÁBITOS:
{habitos}

NIVEL DE ENERGÍA:
{energia}

INSTRUCCIONES:
- Prioriza tareas importantes.
- Evita sobrecarga diaria.
- Agrupa tareas similares.
- Mantén equilibrio entre trabajo, estudio y descanso.

Devuelve:
1. Plan semanal estructurado por día.
2. Justificación breve de la organización.
"""
```

```python
def generar_plan(horarios_fijos, horarios_flexibles, habitos, energia):

    prompt = BASE_PROMPT.format(
        horarios_fijos=horarios_fijos,
        horarios_flexibles=horarios_flexibles,
        habitos=habitos,
        energia=energia
    )

    response = model.generate_content(
        prompt,
        generation_config={
            "temperature": 0.3
        }
    )

    return response.text
```

```python
def replanificar(plan_original, cambio):

    prompt = f"""
Actúa como asistente organizativo adaptativo.

PLAN ORIGINAL:
{plan_original}

CAMBIO IMPREVISTO:
{cambio}

Reorganiza la semana manteniendo prioridades y evitando sobrecarga.
Devuelve solo el nuevo plan actualizado.
"""

    response = model.generate_content(
        prompt,
        generation_config={
            "temperature": 0.2
        }
    )

    return response.text
```
